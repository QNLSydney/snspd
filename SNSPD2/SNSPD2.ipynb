```python
# Oscilloscope count function using search
#self.add_parameter(name="search_total", 
#                           label= "search crossings",
#                           get_cmd="SEARCH:SEARCH1:TOTal?",
#                          unit="counts",)

def count_rate(MS,bin, total):
    MS.clear_search()
    MS.write("TRIGger:A:LEVel:CH1 1")
    start = time.time()
    end = start + total
    last =  int(MS.search_total())

    times = []
    counts= []

    while time.time() < end:
        time.sleep(bin)
        now = time.time()
        count =  int(MS.search_total())
        delta = count 
        counts.append(delta)
        end1 = sum(counts)
    return counts, end1

counts = []
times, counts = count_rate(
    MS = MS, 
    bin = 1, 
    total = 10.01,
)
```

```python
# Calibrating PM beam splitter in corrected configuration
import matplotlib.pyplot as plt

meas = Measurement()
meas.register_custom_parameter("times", label="Samples (approx. s)")
meas.register_custom_parameter("p2", label="W")
meas.register_custom_parameter("p1", label="W")
meas.register_custom_parameter("attenuation", label="dB")

times = []
p1_fiber = []
p2_fiber = []
attenuation_fiber = []
split_low = []
split_high = []

start = time.perf_counter()
with meas.run() as datasaver:
    print(datasaver.run_id)
    for i in range (61):
        t=i*0.5
        p1 = pm120.power() # laser 
        p2 = powermeter.power() # attenuator output 
        split = p1/(p1+p2)
        split2 = p2/(p1+p2)
        split3 = split + split2
        p = 10*np.log10((0.1/0.9*p1)/p2) # attenuation 
        
        times.append(t)
        p1_fiber.append(p1)
        p2_fiber.append(p2) 
        split_low.append(split2)
        split_high.append(split)
        attenuation_fiber.append(p) # attenuation 
        time.sleep(0.5)
        
    datasaver.add_result(("times",times),
                         ("p1", p1_fiber),
                        ("p2", p2_fiber),
                        ("attenuation", attenuation_fiber))

    time.sleep(0.5)

end = time.perf_counter()
print(end-start)
print(split, split2, split3)
plt.plot(times, attenuation_fiber, '.')
plt.plot(times, split_low, '.')
plt.plot(times, split_high, '.')
plt.show()
```
