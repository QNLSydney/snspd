```python
v = 2
p_att.write(f'VOLT {v}')
print(f'Voltage set to {p_att.ask('VOLT?')}')
```

```python
############################ TURN LASER ON ############################ 
laser.enable(True)
print(f'Laser enable status: {laser.enable()}')
time.sleep(10)

voltages = np.arange(0, 7.05, 0.05) # <- same voltage range as Measurement 4-1 for comparison

for v in voltages: 
    
    p_att.write(f'VOLT {v}')
    time.sleep(2)
    print(f'Voltage set to {p_att.ask('VOLT?')}')
    
    calibrate(bs10=params.bs10, bs90=params.bs90, pmeter10=pm100usb, pmeter90=pms120, t=10, attenuator_name=params.att_blue_name, station=station)


############################ TURN LASER OFF ############################ 
laser.enable(False)
print(f'Laser enable status: {laser.enable()}')

print('finished!')
```
